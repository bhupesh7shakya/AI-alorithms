```python
# lab one 
# video reference 
# https://www.youtube.com/watch?v=qul0f79gxGs&list=PLxCzCOWd7aiHGhOHV-nwb0HR5US5GFKFI&index=5
from queue import Queue


def bfs(adj_list):
#     initializing default value
    visited={}
    level={}
    parent={}
    bfs_result=[]
    queue =Queue()

    for node in adj_list.keys():
        visited[node]=False
        parent[node]=None
        level[node]=-1
        
# initialing ends

# as the first index start from A from below example can be start from anything can be customize
#   assume to value to be set a  
    intialized_index=list(adj_list.keys())[0]
#     index of "a" is set be true as it has been visited
    visited[intialized_index]=True
#    index of "a" is set be 0 as it has been main root
    level[intialized_index]=0
# adding "a" to queue
    queue.put(intialized_index)


# checks if the queue was empty 
# incase if it was not empty it be iterate in below code
    while not queue.empty():
    #      pop the parent_node for queue and return first value
        #value of the parent node will be "a" from queue
        parent_node=queue.get()
        #inerting to bfs_result for the traversal value "a"
        bfs_result.append(parent_node)
        #iterate over if adj list with with child nodes
        #"a" contains many child such as b,c or anything
#       #so now it will visit each node too
        for child_node in adj_list[parent_node]:
            #on default visited is false will be updated if it is once visted
            #let assume b is the child of a
            #then now it checks it exits or not 
            # it is really important to have this condition
            # as every node is connected to parent with child and child with parent
            # without the condition it might lead to infinite loop
            if not visited[child_node]:
                #set the not visited node to true for b
                visited[child_node]=True
                # child node parent is parent node ie a is parent b is child of it
                parent[child_node]=parent_node
                #level of child node b is increment of b
                level[child_node] = level[parent_node]+1
                #put this child node to be to repeat 
                #because b might unvisited child node too
                queue.put(child_node)

    print(bfs_result)
    print(level)
    print(parent)
```

```python
adj_list = {
    "A":["B","C","D"],
    "B":["A","E","F"],
    "C":["A","G","H"],
    "D":["A","I"],
    "E":["B","J","K"],
    "F":["B"],
    "G":["C","L"],
    "H":["C"],
    "I":["D","M"],
    "J":["E"],
    "K":["E","N"],
    "L":["G"],
    "M":["I"],
    "N":["N"]
}
bfs(adj_list)
```
